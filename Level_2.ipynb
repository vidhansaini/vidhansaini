# Task 1

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
table_booking_percentage = df['Has Table booking'].value_counts(normalize=True) * 100
online_delivery_percentage = df['Has Online delivery'].value_counts(normalize=True) * 100
online_delivery_percentage
```

```python
table_booking_percentage
```

```python
avg_rating_table_booking = df.groupby('Has Table booking')['Aggregate rating'].mean()
avg_rating_table_booking
```

```python
delivery_price_range = df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack()
delivery_price_range
```

# Task 2

```python
common_price_range = df['Price range'].mode()[0]
common_price_range
```

```python
avg_rating_price_range = df.groupby('Price range')['Aggregate rating'].mean()
avg_rating_price_range
```

```python
best_color = df.groupby('Price range')['Aggregate rating'].mean().idxmax()
color_of_best_rating = df[df['Price range'] == best_color]['Rating color'].unique()
color_of_best_rating
```
